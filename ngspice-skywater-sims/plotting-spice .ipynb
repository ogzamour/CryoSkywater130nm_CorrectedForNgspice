```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#single sweep VGS
data = np.loadtxt("mosfetVGS.txt")
vg = data[:,0]
idrain = -data[:,3]  

plt.figure(figsize=(10,6))
plt.plot(vg, idrain * 1e6)

plt.title("Sky130 NMOS Id–Vgs")
plt.xlabel("VGS (V)")
plt.ylabel("Id (uA)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#single sweep VDS
dataDS = np.loadtxt("mosfetVDS.txt")
vd = dataDS[:,0]
idd = -dataDS[:,3]   

plt.figure(figsize=(10,6))
plt.plot(vd, idd * 1e6)

plt.title("Sky130 NMOS Id–Vds")
plt.xlabel("VDS (V)")
plt.ylabel("Id (uA)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# VGS sweep for variable VDS
steps = ['0', '1', '2', '3', '4', '5']
plt.figure(figsize=(10,6))


for item in steps:

    data1 = np.loadtxt(f"vgs_sweep_step{item}.txt")
    
    # columns: [Vgate, I(VDD)]
    vg1 = data1[:,0]
    idrain1 = data1[:,3]   # sign correction
    vd1 = data1[0,5]
    #plt.figure(figsize=(10,6))
    plt.plot(vg1, idrain1 * 1e6, label= f'VDS = {vd1}')
plt.xlim(0, 1.85)                              # Forces the view limit up to 1.85
plt.xticks(np.arange(0, 1.9, 0.2))

plt.title("Sky130 NMOS Id–Vgs")
plt.xlabel("VGS (V)")
plt.ylabel("Id (uA)")
plt.legend()
plt.grid(True)
    
plt.tight_layout()
plt.show()

#VDS sweep variable VGS
dSteps = ['0', '1', '2', '3', '4']
plt.figure(figsize=(10,6))


for step in dSteps:
    DS = np.loadtxt(f"vds_sweep_step{step}.txt")
    
    
    vd3 = DS[:,0]
    id3 = DS[:,3]   
    vg3 = DS[0,1]
    
    
    plt.plot(vd3, id3 * 1e6, label = f'VGS = {vg3}' )
plt.xlim(0, 1.85)                              # Forces the view limit up to 1.85
plt.xticks(np.arange(0, 1.9, 0.2))
plt.title("Sky130 NMOS Id–Vds")
plt.xlabel("VDS (V)")
plt.ylabel("Id (uA)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
